```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
import random
from skimage.measure import LineModelND, ransac
```

```python
def distance(point1, point2):
    return np.linalg.norm(point1 - point2)

def distance_pitagorath(x1,y1,x2,y2):
    return np.sqrt((x1-x2)**2 + (y1-y2)**2)
    
def parse_polygon_annotation(annotation_line, img_width, img_height):
    parts = list(map(float, annotation_line.split()))
    class_id = int(parts[0])
    points = [(int(x * img_width), int(y * img_height)) for x, y in zip(parts[1::2], parts[2::2])]
    points = np.array(points)
    return class_id, points

def find_closest_points(contour_outer, contour_inner):
    min_distance = float('inf')
    closest_outer_point = None
    closest_inner_point = None

    # 랜덤하게 바깥의 점 하나 선택
    random_outer_point = contour_outer[np.random.randint(len(contour_outer))]

    for point_inner in contour_inner:
        dist = distance(random_outer_point, point_inner)
        if dist < min_distance:
            min_distance = dist
            closest_outer_point = random_outer_point
            closest_inner_point = point_inner

    return closest_outer_point, closest_inner_point

# 바깥 contour와 안쪽 contour의 점들을 받아서 하나의 폴리곤으로 만들어주는 함수
def connect_contours(contour_outer, contour_inner):
    closest_outer_point, closest_inner_point = find_closest_points(contour_outer, contour_inner)
    if closest_outer_point is None or closest_inner_point is None:
        return None
    polygon = np.concatenate((np.flip(contour_outer, axis=0), contour_inner))  # 바깥 contour는 반시계방향으로 배열하고, 안쪽 contour는 시계방향으로 배열한 후 연결

    return polygon


def line_fitting(x_data, y_data):

    def plot_ransac_revised(segment_data_x, segment_data_y):
        data = np.column_stack([segment_data_x, segment_data_y])

        # fit line using all data
        model = LineModelND()
        model.estimate(data)

        # robustly fit line only using inlier data with RANSAC algorithm
        model_robust, inliers = ransac(data, LineModelND, min_samples=2,
                                        residual_threshold=5, max_trials=1000)
        outliers = inliers == False

        # generate coordinates of estimated models
        line_x = np.array([segment_data_x.min(), segment_data_x.max()])
        line_y = model.predict_y(line_x)
        line_y_robust = model_robust.predict_y(line_x)
        k = (line_y_robust[1] - line_y_robust[0]) / (line_x[1] - line_x[0])
        m = line_y_robust[0] - k * line_x[0]
        x0 = (segment_data_y.min() - m) / k
        x1 = (segment_data_y.max() - m) / k
        line_x_y = np.array([x0, x1])
        line_y_robust_y = model_robust.predict_y(line_x_y)
        if (distance_pitagorath(line_x[0], line_y_robust[0], line_x[1], line_y_robust[1]) <
                distance_pitagorath(line_x_y[0], line_y_robust_y[0], line_x_y[1], line_y_robust_y[1])):
            #         plt.plot(line_x, line_y_robust, '-b', label='Robust line model')
            line_twopoint = (line_x, line_y_robust)
        else:
            #         plt.plot(line_x_y, line_y_robust_y, '-b', label='Robust line model')
            line_twopoint = (line_x_y, line_y_robust_y)

        return inliers, outliers, line_twopoint

    x_tmp = x_data.copy()
    y_tmp = y_data.copy()
    ransac_line = []
    while True:
        inliers, outliers, line_twopoint = plot_ransac_revised(x_tmp, y_tmp)

        if x_tmp[inliers].shape[0] >= 2:
            # inliers, two points for line 기록 저장
            ransac_line.append((x_tmp[inliers], y_tmp[inliers], line_twopoint))

        # 나머지 점들 (outliers)
        x_tmp = x_tmp[outliers]
        y_tmp = y_tmp[outliers]

        if x_tmp.shape[0] <= 2 or len(ransac_line) == 4:
            break

    return ransac_line


def line_intersection_points(x_data, y_data, fitted_lines):

    def line_intersection(line1, line2, x_min, x_max, y_min, y_max):
        xdiff = (line1[0][0] - line1[1][0], line2[0][0] - line2[1][0])
        ydiff = (line1[0][1] - line1[1][1], line2[0][1] - line2[1][1])

        def det(a, b):
            return a[0] * b[1] - a[1] * b[0]

        div = det(xdiff, ydiff)
        if div == 0:
            raise Exception('lines do not intersect')

        d = (det(*line1), det(*line2))
        x = det(d, xdiff) / div
        y = det(d, ydiff) / div

        # 범위 내의 값인지 체크
        if x_min -100 <= x <= x_max+ 100 and y_min - 100 <= y <= y_max + 100:
            return x, y
        else:
            return -12345, -12345

    intersection_points = []
    x_min, x_max, y_min, y_max = x_data.min(), x_data.max(), y_data.min(), y_data.max()
    for i in range(len(fitted_lines)):
        for j in range(i+1, len(fitted_lines)):
            (x1, x2), (y1, y2)= fitted_lines[i][2]
            (x3, x4), (y3, y4)= fitted_lines[j][2]
            x, y = line_intersection([[x1, y1], [x2, y2]], [[x3, y3], [x4, y4]], x_min, x_max, y_min, y_max)
            if x != -12345 or y != -12345:
                intersection_points.append((x,y))
    return np.array(intersection_points)
```

```python
image_path = "../SCIE/EIFS_Donut_Only_ribbon/train/1.jpg"
annotation_path = "../SCIE/EIFS_Donut_Only_ribbon/train/annotations_YOLO_only_ribbon/1.txt"
image = cv2.imread(image_path)
img_height, img_width = image.shape[:2]

with open(annotation_path, 'r', encoding='utf-8') as file:
    line = file.readlines()

class_id_outer, polygon_points_outer = parse_polygon_annotation(line[1], img_width, img_height)
class_id_inner, polygon_points_inner = parse_polygon_annotation(line[0], img_width, img_height)

polygon = connect_contours(polygon_points_outer, polygon_points_inner)
```

```python
x_data = polygon_points_outer[:,0]
y_data = polygon_points_outer[:,1]
```

```python
fitted_lines = line_fitting(x_data, y_data)
```

```python
intersection_points = line_intersection_points(x_data, y_data, fitted_lines)
print("교차점:", intersection_points)
```

```python
x_coords = [intersection_points[:, 0]]
y_coords = [intersection_points[:, 1]]

# 이미지 확인
plt.scatter(x_coords, y_coords, color='red', s=200)
plt.plot(polygon[:, 0], polygon[:, 1], 'b-')  # 폴리곤을 파란색 선으로 표시
plt.fill(polygon[:, 0], polygon[:, 1], color='skyblue', edgecolor='blue')
plt.imshow(image)
plt.show()
```
